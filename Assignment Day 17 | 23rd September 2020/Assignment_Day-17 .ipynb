## Question 1:
### ● Using the given MT cars dataset you need to find outlier with z score.

```python
import numpy as np
import pandas as pd
```

```python
mtcars = pd.read_csv("mtcars.csv")
mtcars
```

```python
mtcars.head()
```

```python
mtcars.shape
```

```python
#Mean and std deviation of column [hp]
mtcars_hp=list(mtcars["hp"])
mtcars_hp_mean=np.mean(mtcars_hp)
print("Mean is ", mtcars_hp_mean)
```

```python
mtcars_hp_std=np.std(mtcars_hp)
print("Standard deviation is ", mtcars_hp_std)
```

```python
threshold = 2
outlier = [] 
for i in mtcars["hp"]: 
    z = (i-mtcars_hp_mean)/mtcars_hp_std
    if z > threshold: 
        outlier.append(i) 
print('outlier in dataset is', outlier)
```

```python
import seaborn as sns
sns.boxplot(data=mtcars,x=mtcars['hp'])
```

## Question 2:
### ● Using the Boston data set from sklearn, find out the Outlier and treat them.

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
x = boston.data
y = boston.target
columns = boston.feature_names

#create the dataframe
boston_df = pd.DataFrame(boston.data)
boston_df.columns = columns
boston_df.head()
```

```python
boston_df.shape
```

```python
boston_df.describe()
```

```python
import seaborn as sns
sns.boxplot(data=boston_df,x=boston_df['DIS'])
```

```python
Q1 = boston_df['DIS'].quantile(0.25)
Q3 = boston_df['DIS'].quantile(0.75)

IQR = Q3 - Q1

print(Q1)
print(Q3)
print(IQR)
```

```python
lower_value = Q1 - (1.5 * IQR)
upper_value = Q3 + (1.5 * IQR)

print(lower_value, upper_value)
```

```python
for i in boston_df["DIS"]:
    if i > upper_value:
        print("Outliers are : ",i)
    elif i < lower_value:
        print("Outliers are : ",i)
```

```python
boston_df = boston_df[boston_df['DIS'] < upper_value]
boston_df.head()
```
